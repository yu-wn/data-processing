```python
# フォルダ内すべてのcsvファイルについて,センサ値を力に変換(3回目)

import sys
import numpy as np
import glob
import os
import pandas as pd

# データの保存先ディレクトリを指定
save_path = r"E:\てすりとキネクト（てすり）\3回目変換後"

# CSVファイルが格納されているディレクトリを指定
dataList = glob.glob(r"E:\てすりとキネクト（てすり）\3回目\*.csv")  # フォルダ内のすべてのCSVファイルを取得

# キャリブレーション変換行列M（北村さんが計算して下さった）
M = np.array([[0.15086421408741382, 0.006742702042989319, -0.004906646648369396, -151.31438216515633],
              [0.0031942886490321796, 0.09206720241633937, 0.012697342020777776, -147.420102232898],
              [-0.010719892902014022, 0.007780541000191522, 0.13143420163825484, -71.99949755799366],
              [3.686287386450715e-18, 3.5236570605778894e-19, -6.098637220230962e-19, 0.9999999999999974]])

# データの変換と保存
for fileName in dataList:
    fileData = open(fileName, 'r')
    
    # 元のファイル名に "_force3.csv" を追加して保存ファイル名を作成
    outputFileName_x = os.path.join(save_path, os.path.basename(fileName).replace(".csv", "_force3.csv"))
    
    # 保存用のファイルを開く
    fo = open(outputFileName_x, "w")

    for line in fileData:  # 読み込んだファイルの先頭から一行ずつ処理
        cells = line[:-1].split(",")  # 最後の改行コード以外を","で区切ってリストにする

        # センサーIDが1の場合
        if cells[1] == "1":
            sensorData = [float(cells[2]), float(cells[3]), float(cells[4])]
            print(sensorData)
            # 力の変換
            x = sensorData[0]*M[0,0] + sensorData[1]*M[0,1] + sensorData[2]*M[0,2] + M[0,3]
            y = sensorData[0]*M[1,0] + sensorData[1]*M[1,1] + sensorData[2]*M[1,2] + M[1,3]
            z = sensorData[0]*M[2,0] + sensorData[1]*M[2,1] + sensorData[2]*M[2,2] + M[2,3]
            
            # 変換後のデータを保存
            fo.write("%s,%s,%lf,%lf,%lf\n" % (cells[0], cells[1], x, y, z))

    fileData.close()
    fo.close()
```
